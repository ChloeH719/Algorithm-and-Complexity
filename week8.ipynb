```python
import pandas as pd
import numpy as np

class BinaryLogisticRegression:
    def __init__(self, learning_rate=0.01, epochs=1000):
        # Constructor to initialize the learning rate and number of epochs
        self.learning_rate = learning_rate  # The step size used for each update
        self.epochs = epochs  # The number of times to iterate through the training set
        self.weights = None  # Weights of the model, to be defined later
        self.bias = None  # Bias of the model, to be defined later

    # sigmoid function
    def sigmoid(self, z):
        # maps any real value into the range [0, 1]
        return 1 / (1 + np.exp(-z))

    # cost function
    def loss(self, y, t):
        # computes the loss for predicted y vs actual target t
        return 0.5 * (y - t)**2

    def predict(self, X):
        linear_model = np.dot(X, self.weights) + self.bias  # Linear part of the hypothesis
        y_predicted = self.sigmoid(linear_model)  # Apply sigmoid activation
        return y_predicted

    def fit(self, X, t):
        # Train the logistic regression model using stochastic gradient descent
        n_samples, n_features = X.shape  # Get the number of samples and features
        self.weights = np.zeros(n_features)  # Initialize weights as a zero vector
        self.bias = 0  # Initialize bias to zero

        # Stochastic Gradient Descent
        for _ in range(self.epochs):
            for idx, x_i in enumerate(X):
                # For each sample in the dataset, compute the prediction and update weights and bias
                linear_model = np.dot(x_i, self.weights) + self.bias  # Compute linear model
                y_predicted = self.sigmoid(linear_model)  # Apply sigmoid to get prediction
                
                # Compute the gradient of loss wrt weights and bias
                dw = (y_predicted - t[idx]) * y_predicted * (1 - y_predicted) * x_i
                db = (y_predicted - t[idx]) * y_predicted * (1 - y_predicted)

                # Update the weights and bias
                self.weights -= self.learning_rate * dw
                self.bias -= self.learning_rate * db

# Load dataset
df = pd.read_csv('mushrooms.csv')

# Each categorical feature value is turned into a new binary column
df = pd.get_dummies(df)

# Define features and target variable
X = df.drop('class_p', axis=1).values
t = df['class_p'].values

# Initialize and train the logistic regression model
model = BinaryLogisticRegression(learning_rate=0.01, epochs=1000)
model.fit(X, t)

# Predict the class (poisonous or not) for the training data
predictions = model.predict(X)
predictions = [1 if p >= 0.5 else 0 for p in predictions]

# Print the accuracy
accuracy = np.sum(predictions == t) / len(t)
print("Accuracy:", accuracy)
```
